## Read a PDF with `read_pdf()` function

```python
#Eextracting the tables from the original pdf and storing in dataframes
import tabula
pdf_path = "/content/OScoursepolicy.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True, pages ="all")
# read_pdf returns list of DataFrames
print(len(dfs))
dfs[0]
```

```python
#Creating .csv file and storing each table individually
for i in range(0,len(dfs)):
    dfs[i].to_csv(f"Tabula_file{i}.csv")
print("Individual Files of Tables created successfully")
```
