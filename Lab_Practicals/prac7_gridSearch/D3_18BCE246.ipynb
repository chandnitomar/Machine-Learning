Loading Dataset - iris

```
from sklearn import neighbors, datasets, metrics, svm
from sklearn.model_selection import GridSearchCV

X,y = datasets.load_iris(return_X_y = True) 

X_train = X[range(1,150,2),:]
y_train = y[range(1,150,2)]

#print(y_train.shape)
#print(X_train.shape)

X_test = X[range(0,150,2),:]
y_test = y[range(0,150,2)]
```

KNN classifier gridsearch

```
clf = neighbors.KNeighborsClassifier()
test_hyperparam = [{'n_neighbors': [2,5,6,7,14,17],'weights': ['uniform','distance']}]

clf_gridsearch = GridSearchCV(clf,test_hyperparam,cv = 5,refit = True)
clf_gridsearch.fit(X_train,y_train)

print("BestParameters are #neighbors="+str(clf_gridsearch.best_params_['n_neighbors'])+" and weights="+clf_gridsearch.best_params_['weights'])
y_pred = clf_gridsearch.predict(X_test)
print(y_pred)

print("Accuracy: ",metrics.accuracy_score(y_test,y_pred,normalize=True))
print("\nClassification Report:\n",metrics.classification_report(y_test,y_pred))
print("\nConfusion Matrix: \n",metrics.confusion_matrix(y_test,y_pred))

```

```
clf = neighbors.KNeighborsClassifier(n_neighbors= clf_gridsearch.best_params_['n_neighbors'],weights = clf_gridsearch.best_params_['weights'])
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

print(y_pred)

print("Accuracy: ",metrics.accuracy_score(y_test,y_pred,normalize=True))
print("\nClassification Report:\n",metrics.classification_report(y_test,y_pred))
print("\nConfusion Matrix: \n",metrics.confusion_matrix(y_test,y_pred))



            
```

## Support Vector Machine

```
classifier = svm.SVC()
classifier.fit(X_train, y_train)
y_predicted = classifier.predict(X_test)
print('Predicted Values : ', y_predicted)
```

```
print("Accuracy: ",metrics.accuracy_score(y_test,y_predicted,normalize=True))
print("\nClassification Report:\n",metrics.classification_report(y_test,y_predicted))
print("\nConfusion Matrix: \n",metrics.confusion_matrix(y_test,y_predicted))


```
