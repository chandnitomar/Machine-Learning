## **Reading text from Image using pytesseract and tesseract-ocr**

```
!pip install pytesseract
!sudo apt install tesseract-ocr
```

```
import pytesseract as tess
try:
    import Image

except ImportError:
    from PIL import Image

 #Changing the directory as the pytesseract library is only wrapper and the real engine has to be referenced seperately
tess.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

#Displaying the Image to process
img = Image.open(r'/content/OCR.jpg')
print("Original Image(OCR.jpg)")
display(img)
```

```
#Extracting the text from Image(OCR.jpg)
text = tess.image_to_string(img)
print(text)
```
